# Supply Chain Data Analytics
## Phase 1 – Data Cleaning & Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Dataset Loading
Loading the supply chain dataset into the Python environment.

```python
df = pd.read_csv(
    "DescriptionDataCoSupplyChain.csv",
    encoding="latin1",
    encoding_errors="ignore",
    on_bad_lines="skip"
)

df.head()
```

## Dataset Overview
Understanding dataset dimensions and structure.

```python
print(df.shape)
print(df.columns)
```

## Missing Value Analysis
Identifying null values present in the dataset.

```python
df.isnull().sum()
```

## Data Cleaning & Preprocessing
Handling missing values, duplicate records, and formatting inconsistencies.

```python
# Remove fully empty columns
df = df.dropna(axis=1, how='all')

# Remove columns with excessive missing values
threshold = len(df) * 0.7
df = df.dropna(thresh=threshold, axis=1)

# Fill missing values
for col in df.columns:

    if df[col].dtype != 'object':
        df[col] = df[col].fillna(df[col].median())

    else:
        df[col] = df[col].fillna(df[col].mode()[0])

# Remove duplicates
df = df.drop_duplicates()
```

## Data Preprocessing
Standardizing text values and preparing the dataset for analysis.

```python
for col in df.select_dtypes(include='object'):
    df[col] = df[col].astype(str).str.strip()
```

## Cleaned Dataset Information
Summary of cleaned dataset structure and statistics.

```python
print(df.info())
print(df.describe())
```

## Exploratory Data Analysis (EDA)
Visualizing important dataset patterns and distributions.

```python
df['Sales'].hist()
plt.show()
df['Order Item Quantity'].hist()
plt.show()
```

## Saving Cleaned Dataset
Exporting the cleaned dataset for further analysis and dashboard development.

```python
df.to_csv("cleaned_DescriptionDataCoSupplyChain.csv", index=False)
```
